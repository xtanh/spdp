```python
import os
import csv
import time
import json
import torch
import random
import argparse
import numpy as np
import pandas as pd
from tqdm.auto import tqdm

from model import DDG_RDE_Network, Codebook
from trainer import CrossValidation, recursive_to
from dataloader import SkempiDatasetManager
from utils import set_seed, check_dir, eval_skempi_three_modes, save_code, load_config

from ipdb import set_trace


def pretrain(epoch, model, optimizer, dataloader):
    model.train()
    
    for batch_idx, batch in enumerate(dataloader):
        if batch is None:
            continue
        batch = recursive_to(batch, device)
        if batch is None:
            print(f"Error: Batch is None at epoch {epoch}")
            return
        _, e_q_loss, s_recon_loss, h_recon_loss, x_recon_loss = model(batch)
        loss_vae = e_q_loss * 0.005 + s_recon_loss + h_recon_loss + x_recon_loss
        print(loss_vae)
        # 继续训练逻辑
    print(f"Completed epoch {epoch}")

```

```python
save_dir = '/home/hongtan/pretrain/weight'
param = json.loads(open("/home/hongtan/pretrain/config/param_configs.json", 'r').read())
args = argparse.Namespace(**param)
vae_model = Codebook(args).to(device)
vae_optimizer = torch.optim.Adam(vae_model.parameters(), lr=args.lr, weight_decay=args.weight_decay)

dataloader_manager = SkempiDatasetManager(batch_size=16)

for epoch in range(1, args.pre_epoch+1):
    dataloader = dataloader_manager.get_pretrain_loader()  # 每个 epoch 获取新的 DataLoader
    pretrain(epoch, vae_model, vae_optimizer, dataloader)

for p in vae_model.parameters():
    p.requires_grad_(False)

torch.save(vae_model.state_dict(), os.path.join(save_dir, 'checkpoint', f'vae_model.ckpt'))
```

```python
import pandas as pd

# 读取CSV数据
data = pd.read_csv('/home/hongtan/pretrain_single/data/modified_file.csv')

# 创建一个列表，用于保存突变前后氨基酸一致的行
same_mutations = []

# 遍历每一行数据
for index, row in data.iterrows():
    # 获取突变前的氨基酸和突变后的氨基酸
    mutation = row['corrected mutations']
    wt_aa = mutation[0]  # 突变前的氨基酸
    mt_aa = mutation[-1]  # 突变后的氨基酸
    
    # 检查突变前后的氨基酸是否相同
    if wt_aa == mt_aa:
        same_mutations.append(row)

# 将结果转换为 DataFrame 以便查看
same_mutations_df = pd.DataFrame(same_mutations)

# 输出结果
if not same_mutations_df.empty:
    print("突变前后氨基酸一致的行：")
    print(same_mutations_df)
else:
    print("没有突变前后氨基酸一致的行。")
```
